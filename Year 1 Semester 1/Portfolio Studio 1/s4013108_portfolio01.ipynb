What problem is ideation intended to prevent?

*It is intended to prevent the problem of users experiencing their pain points and/or help them achieve their goals with the product using different ideas or concepts of different ideas as potential solutions to the problem. Optimal ideation techniques also help prevent the occurence of groupthink while brainstorming ideas, where a couple of people in the group leads the conversation about the ideas, which limits the ability of other members from speaking out and contributing, which reduces creativity overall*

**Question 2:**  <a id="q-sc3-q2"></a>
Describe each of the 5 stages of the design charrette ideation technique in 1
sentence.


| Stage Name | Description |
|------------|-------------|
| *Stage 1* | *Inform the problem that the team is trying to solve as so as current knowledge of gathered user experience data of pain points, goals, and personas to the whole team* |
| *Stage 2* | *Make every member of the team sketch/ideate/conceptualize one or more product solutions to the pain points* |
| *Stage 3* | *Share these ideas to the entire team and see how well each idea solves the pain points, while also letting other members critique or ask questions about the idea and taking note of the unique concepts of each member's idea, as so as voting on which idea each member likes best* |
| *Stage 4* | *Make each member sketch/ideate/conceptualize new or improved upon ideas in pairs or small groups to better discuss and justify their solutions to their colleagues and build upon each other's creativity* |
| *Stage 5* | *Share these ideas again to the entire team, along with critiquing and questioning each idea and seeing which one the team likes best, which would hopefully narrow down the search for the ideas that the team will build upon into prototypes in the future* |

`````python
package prereq;

import java.util.ArrayList;
import java.util.Scanner;

public class WordFrequencies {
   public static void main(String[] args) {  
      Scanner scnr = new Scanner(System.in);
      int num = 0;  
      ArrayList<String> strArr = new ArrayList<String>();  
      // STEP 1: Read the number of words into an integer
      num = scnr.nextInt();

      // STEP 2: Read the words, one-at-a-time into an ArrayList
      for (int i = 0; i < num; ++i) {
         strArr.add(scnr.next());
      }

      // STEP 3: Find the number of times the *first* word appears in the ArrayList.
      //         Print out this number in the required format.
      int countf = 0;
      for (int i = 0; i < num; ++i) {
         if (strArr.get(i).equals(strArr.get(0))) {
            ++countf;
         }
      }
      System.out.println(strArr.get(0) + " - " + countf);

      // STEP 4: Question - What do you need to change in your Step 3 code
      //                    to find the number of times the *second* word
      //                    appears in the ArrayList?
                              // change the indexes from 0 to 1

      //         Question - What about the *third* word?
                              // change the indexes from 0 to 2
      //         Question - Do you see a pattern emerging?
                              // yes, change the index to the index of the string you want to find the frequency of

      // STEP 5: Turn your code in Step 3 into a loop that checks every word in the ArrayList

      for (int i = 0; i < num; ++i) {
         int count = 0;
         for (int j = 0; j < num; ++j) {
            if (strArr.get(j).equals(strArr.get(i))) {
               ++count;
            }
         }
         System.out.println(strArr.get(i) + " - " + count);
      }
   }
}
`````
