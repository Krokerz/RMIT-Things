List 3 benefits of normalisation

*Minimizes the replication or duplication of redundant data; Makes the data easier to find, edit, and maintain; Allow for performing operations without generating garbage inside of it or invalidating the state of it*

**Studio Class (SC) 15, Question 2:** <a id="q15b"></a>

Activity - Describe in one sentence (in your own words)

|Question | Answer |
|---|---|
| 1NF | *A relation that only has a singular value for each attribute, does not have multiple columns that hold the same kind of data, and that it has a primary key* | 
| 2NF | *A relation that has all the properties of 1NF and has a singular attribute for the primary key or a multi-attribute primary key that also has the attributes that are not primary keys be reliant on the multi-attribute primary key* |
| 3NF | *A relation that has all the properties of 2NF and has all of the attributes that are not primary keys not be reliant on any of the other attributes that are not primary keys* |
